```python
import time
from transformers import pipeline

# ==========================================
# REAL-TIME TRANSLATOR (German -> English)
# ==========================================

print("Loading AI Model... (This happens only once)")

# 1. Load Pre-trained Model
# "Helsinki-NLP/opus-mt-de-en" is a state-of-the-art model for German->English
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-de-en")

print("\n=== REAL-TIME AI TRANSLATOR ===")
print("Type a German sentence and press Enter. Type 'q' to quit.\n")

while True:
    # 2. Get Input
    try:
        user_input = input("🇩🇪 German: ")
    except KeyboardInterrupt:
        break

    if user_input.lower() in ['q', 'quit', 'exit']:
        print("Exiting...")
        break

    if not user_input.strip():
        continue

    # 3. Real-Time Inference
    start_time = time.time()

    # The actual AI translation happens here
    outputs = translator(user_input)
    translated_text = outputs[0]['translation_text']

    end_time = time.time()

    # 4. Output Result
    print(f"🇺🇸 English: {translated_text}")
    print(f"⏱️  Latency: {(end_time - start_time):.2f} seconds\n")
```
